```python
from numpy import cos, sin, arcsin, sqrt
from math import radians

def haversine(row):
    lat1 = 34.050536
    lon1 = -118.247861
    lat2 = row['Latitude']
    lon2 = row['Longitude']

    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    km = 6371 * c

    return km


dfbradbuild['distance'] = dfbradbuild.apply(lambda row: haversine(row), axis=1)
dfbradbuild
```
